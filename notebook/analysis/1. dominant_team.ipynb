```sql
SELECT
  team_name,
  COUNT(1) AS total_races,
  SUM(calculated_points) AS total_points,
  AVG(calculated_points) AS avg_points
FROM
  f1_presentation.calculated_race_results
GROUP BY team_name
HAVING COUNT(1) > 50
ORDER BY avg_points DESC;
```

```sql
SELECT
  team_name,
  COUNT(1) AS total_races,
  SUM(calculated_points) AS total_points,
  AVG(calculated_points) AS avg_points
FROM
  f1_presentation.calculated_race_results
WHERE race_year BETWEEN 2010 AND 2020
GROUP BY team_name
HAVING COUNT(1) > 50
ORDER BY avg_points DESC;
```
